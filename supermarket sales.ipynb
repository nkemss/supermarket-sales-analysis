```python
df.groupby('Payment')['cogs'].mean()
```

1. The most common payment method is E-wallet, showing a preference for digital transactions.
2. Customers using Cash spend more on average compared to other payment methods.
3. E-wallet payments are frequent but associated with lower average spending.

```python
df.groupby('Payment')['cogs'].sum()
```

```python
sns.barplot(x='Payment',y='cogs',data=df)
```

```python
df.groupby(['Customer type','Payment'])['cogs'].mean()
```

Member customers spend more on average compared to normal customers

```python
df.groupby(['City','Payment'])['cogs'].sum()
```

In Mandalay, Credit card payment generates higher revenue while in Naypyitaw, Cash generates higher revenue and E-wallet generates the most revenue in Yangon. 

```python
df.groupby(['City','Payment'])['cogs'].sum().unstack().plot(kind='bar')
```

```python
df.groupby(['City','Payment'])['cogs'].agg(['sum','count','mean'])
```

1. In Mandalay,although E-wallet is the most commonly used, Credit card payment generates the most revenue.
2. In Naypyitaw, Cash payment is the most common and it also generates the most revenue. But on average, E-wallet is the highest.
3. In Yangon, E-wallet is the most commonly used and also produces the most revenue.
